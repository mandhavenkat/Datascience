###stack plot -- to compare different groups

```python
#example
from matplotlib import pyplot as plt
days=[1,2,3,4,5]
sleeping=[7,8,6,11,7]
eating=[2,3,4,3,2]
working=[7,8,7,2,2]
playing=[8,5,7,8,13]
plt.plot([],[],color='m',label='Sleeping',linewidth=5)  #to add label
plt.plot([],[],color='c',label='Eating',linewidth=5)  # to add label
plt.plot([],[],color='r',label='Working',linewidth=5) # to add label
plt.plot([],[],color='g',label='Playing',linewidth=5) # to add label
            ## days as x , y1,y2,y3,y4,y5, color)
plt.stackplot(days,sleeping,eating,working,playing,colors=['m','c','r','g'])
plt.xlabel('x')
plt.ylabel('y')
plt.title('Stack Plot')
plt.legend()
plt.show()
```

### Pie Chart -- to show the data in %

```python
from matplotlib import pyplot as plt
slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
colr=['c','m','r','b']
# data will be plotted anti clock wise
#explode will move the respective plot area away based on the given value, 
#    explode should have the count matching with labels
#start angle by default starts from 90 where as the value will be given as 0,
#    when coded the value as 90 then it will be equal to 0
plt.pie(slices,
        labels=activities,
        colors=colr,
        startangle=90,
        shadow=True,
        explode=(0.1,0.1,0.4,0),autopct='%1.3f%%')
        #autopct='%1.1f%%')
plt.title('Pie Plot')
plt.show()
           
```

###Sub plots

```python
import numpy as np
from matplotlib import pyplot as plt
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1=np.arange(0.0,5.0,0.1)
t2=np.arange(0.0,5.0,0.2)
print('plot 1')
plt.subplot(211)  #vertically 2 plots horizontally 1 and 1st plot
plt.plot(t1,f(t1),'bo',t2,f(t2))

plt.subplot(212)  #verticallyt 2 plots horizontally 1 and 2nd plot
plt.plot(t2,np.cos(2*np.pi*t2))
plt.show()

#write the code in other way
print('plot 2')
plt.subplot(121)  #vertically 1 plots horizontally 2 and 1st plot
plt.plot(t1,f(t1),'bo',t2,f(t2))

plt.subplot(122)  #verticallyt 1 plots horizontally 2 and 2nd plot
plt.plot(t2,np.cos(2*np.pi*t2))
plt.show()
```

```python
# scatter plot using pandas, parameter s represents the dot size off the respective data points 
print("For Sepal Length and Width comparison")
ax = data[data.Species=='Iris-setosa'].plot.scatter(x='SepalLengthCm', y='SepalWidthCm', s=50,
                                                    color='red', label='setosa')
data[data.Species=='Iris-versicolor'].plot.scatter(x='SepalLengthCm', y='SepalWidthCm', s=50,
                                                color='green', label='versicolor', ax=ax)
data[data.Species=='Iris-virginica'].plot.scatter(x='SepalLengthCm', y='SepalWidthCm',  s=50,
                                                color='blue', label='virginica', ax=ax,)
ax.set_title("scatter")
```
